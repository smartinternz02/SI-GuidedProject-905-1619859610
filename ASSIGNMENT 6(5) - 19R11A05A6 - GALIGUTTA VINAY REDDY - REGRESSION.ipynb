```python
x=dt.iloc[:,0:6].values
y=dt.iloc[:,6].values
```

```python
x.shape
```

```python
y.shape
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
ct=ColumnTransformer([("oh",OneHotEncoder(),[1,4,5])],remainder="passthrough")
#coverts data into binary form

x=ct.fit_transform(x)
```

```python
x.shape
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
x_train
```

```python
y_train
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x_train=sc.fit_transform(x_train)
x_test=sc.fit_transform(x_test)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
ann=Sequential()
```

```python
ann.add(Dense(6,kernel_initializer="random_uniform",activation="relu"))
```

```python
ann.add(Dense(12,kernel_initializer="random_uniform",activation="relu"))
```

```python
ann.add(Dense(1,kernel_initializer="random_uniform",activation="relu"))
```

```python
ann.compile("adam","mean_squared_error",metrics=["accuracy"])
```

```python
ann.fit(x_train,y_train,batch_size=32,epochs=500)
```

```python
ypred=ann.predict(x_test)
```
